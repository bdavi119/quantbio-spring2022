The goal of this lab is to practice simulation tools on the recently instructed topic of binding. Write a notebook that simulates ligand binding for a macromolecule that has two binding sites.

**Do the following:**

A protein has two binding sites. The Kd for binding the first ligand is 10<sup>-6</sup>. Plot the Fraction bound vs. concentration of ligand (log scale) if the cooperativity ($\alpha$) between the first and second binding sites is 
- a) 100 
- b) 1
- c) 0.01

## Libraries you will need

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Student Input Cell

```python
coop = [100, 1, 0.01]
step = 0.05
ligand_log = np.arange(-10, 0+step, step)
ligand = 10**ligand_log
kd = 10 ** (-6)

fig, ax = plt.subplots(figsize=(8,6))

nobound = 1
onebound = 2 * ligand / kd
for i in range(3):
    a = coop[i]
    twobound = ligand ** 2 / (a * (kd ** 2))
    Q = nobound + onebound + twobound
    bound_frac = (0.5 * onebound + twobound) / Q
    
    ax.plot(ligand_log, bound_frac)

plt.title('Double Ligand Binding at Different Cooperativities')
plt.xlabel('log([ligand])')
plt.ylabel('Fraction of binding sites with bound ligand')
plt.legend(['\u03B1 = 100', '\u03B1 = 1', '\u03B1 = 0.1'])
plt.show()
```
