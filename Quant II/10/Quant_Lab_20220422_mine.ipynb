## Libraries you will need

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Question 2a:
Assuming that both steps in this reaction are first-order elementary reactions, plot concentration vs. time for A, I and B for [A]<sub>0</sub> = 1 in the range of 0 ≤ t ≤ 25 seconds, using:

k<sub>1</sub> = 0.25 s<sup>-1</sup> and k<sub>2</sub> = 1 s<sup>-1</sup>. 

## Student Input Cell

```python
step = .1
T = np.arange(0, 25+step, step)

k1 = 0.25
k2 = 1
A0 = 1
A1 = A0 * np.exp(-k1 * T)
I1 = k1 * A0 * (np.exp(-k1 * T) - np.exp(-k2 * T)) / (k2 - k1)
B1 = A0 * (1- (k2 * np.exp(-k1 * T) - k1 * np.exp(-k2 * T)) / (k2 - k1))

plt.plot(T, A1, T, I1, T, B1)
plt.legend(['A', 'I', 'B'])
plt.title('Concentration of Species in the Reaction A \u2192 I \u2192 B\n'+
          'k1 = 0.25, k2 = 1')
plt.xlabel('Time (s)')
plt.ylabel('Concentration (M)')
plt.show()
```

## Question 2b:
Calculate and plot the concentrations for A, I and B for k<sub>1</sub> = 1 s<sup>-1</sup> and k<sub>2</sub> = 0.25 s<sup>-1</sup>. How do the changes in concentration over time compare for each of the three species?

Plot A, I, and B on three separate graphs. For each of these reactants, plot the concentrations for each when k<sub>1</sub> = 1 s<sup>-1</sup> &  k<sub>2</sub> = 0.25 s<sup>-1</sup>, and when k<sub>1</sub> = 0.25 s<sup>-1</sup> &  k<sub>2</sub> = 1 s<sup>-1</sup>

## Student Input Cell

```python
k1 = 1
k2 = 0.25
A0 = 1
A2 = A0 * np.exp(-k1 * T)
I2 = k1 * A0 * (np.exp(-k1 * T) - np.exp(-k2 * T)) / (k2 - k1)
B2 = A0 * (1- (k2 * np.exp(-k1 * T) - k1 * np.exp(-k2 * T)) / (k2 - k1))

plt.plot(T, A2, T, I2, T, B2)
plt.legend(['A', 'I', 'B'])
plt.title('Concentration of Species in the Reaction A \u2192 I \u2192 B\n'+
          'k1 = 1, k2 = 0.25')
plt.xlabel('Time (s)')
plt.ylabel('Concentration (M)')
plt.show()
```

```python
fig, ax = plt.subplots(3, 1, figsize = (9, 8))

species = [('A', A1, A2), ('I', I1, I2), ('B', B1, B2)]

for i in range(3):
    spec = species[i]
    ax[i].plot(T, spec[1], T, spec[2], alpha=0.5)
    ax[i].set_title('Concentration of {} Over Time'.format(spec[0]))
    ax[i].set_xlabel('Time (s)')
    ax[i].set_ylabel('Concentration (M)')

ax[1].legend(['k1 = 0.25, k2 = 1', 'k1 = 1, k2 = 0.25'], loc=7, bbox_to_anchor=(1.2, 0.5), fontsize='x-large')
plt.tight_layout()
plt.show()
```
